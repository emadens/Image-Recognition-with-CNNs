```python
import os
import time 

import random
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
import tensorflow as tf
import numpy as np
import cv2

tfk = tf.keras
tfkl = tf.keras.layers
print(tf.__version__)
```

```python
inc_model = tfk.models.load_model('../input/modelli/sub8/SubmissionModel')
xce_model = tfk.models.load_model('../input/modelli/sub9/SubmissionModel')

xce_model._name = 'modello2'
```

```python
inputs = tfk.Input(shape=(256,256,3))

a = (inputs/255)
b = tf.keras.applications.mobilenet.preprocess_input(inputs)

x = inc_model(a)
y = xce_model(b)

avg = tf.keras.layers.Average()([x, y])


comb_model = tfk.Model(inputs=inputs, outputs=avg, name='Comb_model')


comb_model.summary()
```

```python
comb_model.save('Combined_model')
```
